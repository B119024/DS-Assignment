***Reading the data using Pandas***

```python
df = pd.read_csv('/kaggle/input/sample-data/phase_data_sample.csv',index_col=None)
```

```python
df.dropna(how='any')
```

```python
df.std()
```

```python
x = df.drop('label',axis=1)
y = df['label']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
x_scaled.mean(axis=0)
```

```python
x_scaled.std(axis=0)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,test_size=0.2,random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
std_model = LinearRegression()
std_model.fit(x_train,y_train)
```

```python
y_pred = std_model.predict(x_test)
pd.DataFrame({'y_test': y_test,'y_pred': y_pred})
```

```python
r2score = std_model.score(x_test,y_test)
print(r2score)
```

```python
X = df.iloc[:,[16,83,105,118]]
Y = df['label']
```

```python
print(X)
print(Y)
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled.mean(axis=0)
```

```python
X_scaled.std(axis=0)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X_scaled,Y,test_size=0.2,random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
std_model = LinearRegression()
std_model.fit(X_train,Y_train)
```

```python
Y_pred = std_model.predict(X_test)
pd.DataFrame({'Y_test': Y_test,'Y_pred': Y_pred})
```

```python
r2score = std_model.score(X_test,Y_test)
print(r2score*100,'%')
```

```python
from sklearn.ensemble import RandomForestRegressor
forest_model = RandomForestRegressor()
forest_model.fit(X_train,Y_train)
```

```python
forest_pred = forest_model.predict(X_test)
pd.DataFrame({'y_test': Y_test,'forest_pred': forest_pred})
```

```python
r2score = forest_model.score(X_test,Y_test)
print(r2score)
```
